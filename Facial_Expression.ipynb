```python
import tensorflow as tf
from tensorflow import keras
import os
import cv2
```

```python
model = keras.models.load_model(r'C:\Users\Dell\Downloads\Facial Expression.keras')
print(" Model loaded successfully!")
```

```python

emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

input_h, input_w = model.input_shape[1], model.input_shape[2]
print(f"Model input size: {input_h}x{input_w}")


cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print(" Error: Could not open webcam.")
    exit()

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    ret, frame = cap.read()
    if not ret:
        break

   
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]

      
        resized = cv2.resize(roi_gray, (input_w, input_h))
        normalized = resized / 255.0
        img_input = normalized.reshape(1, input_h, input_w, 1)  # grayscale → 1 channel

    
        prediction = model.predict(img_input)
        label_index = np.argmax(prediction)
        label = emotion_labels[label_index]
        confidence = np.max(prediction)

     
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
        cv2.putText(frame, f"{label} ({confidence*100:.1f}%)", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255,255,255), 2)

    
    cv2.imshow("Facial Expression Detection", frame)

    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```
